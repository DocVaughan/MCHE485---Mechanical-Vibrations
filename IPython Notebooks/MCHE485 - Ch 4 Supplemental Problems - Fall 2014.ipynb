## Problem 4.61

```python
# Define the known parameters
m1 = 1000.0         # kg
k1 = 36000.0        # N/m

F1 = 0.7 # N
F = [F1, 0]

w = 6.0   # rad/s

# Set up the range of m2 masses to look at
mass_range = np.linspace(0.1,1,200)

# Create an appropriately sized array of zeros to fill with amplitudes
x2 = np.zeros_like(mass_range)

# Now's let's see how |x2| changes with m2
for index, m2 in enumerate(mass_range):
    k2 = k1/m1 * m2    # N/m

    M = np.asarray([[m1, 0],
                    [0,  m2]])

    K = np.asarray([[k1 + k2, -k2],
                    [-k2,      k2]])
    
    # This is (K-w^2 M)^-1 * (w^2*F)
    X = np.dot(linalg.inv(K - w**2 * M), F)

    x2[index] = X[1]
```

```python
# Set the plot size - 3x2 aspect ratio is best
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Change the axis units to serif
setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
xlabel(r'$m_2$ mass (kg)',family='serif',fontsize=22,weight='bold',labelpad=5)
ylabel(r'$|x_2|$ (cm)',family='serif',fontsize=22,weight='bold',labelpad=10)

plot(mass_range, x2, linewidth=2, label=r'$|x_2|$')

# uncomment below and set limits if needed
# xlim(0,5)
# ylim(0,10)

# # Create the legend, then fix the fontsize
# leg = legend(loc='upper right', fancybox=True)
# ltext  = leg.get_texts()
# setp(ltext,family='serif',fontsize=18)

# Adjust the page layout filling the page using the new tight_layout command
tight_layout(pad=0.5)

# save the figure as a high-res pdf in the current folder
# It's saved at the original 6x4 size
# savefig('Problem4p61.pdf')

fig.set_size_inches(9,6) # Resize the figure for better display in the notebook
```

```python
# We can also use a numerical solver to get a more exact solution

# Import the optimize module (We'll actually use a minimization function)
from scipy import optimize

def x2_amp_diff(m2):
    k1 = 36000.0
    m1 = 1000.0
    k2 = k1 / m1 * m2[0]    # N/m
    
    M = np.asarray([[m1, 0],
                    [0,  m2]])

    K = np.asarray([[k1 + k2, -k2],
                    [-k2,      k2]])
    
    # This is (K-w^2 M)^-1 * (w^2*F)
    X = np.dot(linalg.inv(K - w**2 * M), F)

    x2 = X[1]
    
    return x2 + 0.1

solution = optimize.fsolve(x2_amp_diff, 2)

print '\nThe second mass should be {:0.4f}kg.\n'.format(solution[0])
    
```

## Problem 4.63

```python
# Define the known parameters
m1 = 10.0         # kg
k1 = 4000.0        # N/m

F1 = 10.0          # N
F = [F1, 0]

w = 20.0           # input frequency (rad/s)

# Vibration Absorber case
m2 = 10.0          # kg
k2 = 4000.0        # N/m

M = np.asarray([[m1, 0],
                [0,  m2]])

K = np.asarray([[k1 + k2, -k2],
                [-k2,      k2]])

# This is (K-w^2 M)^-1 * (w^2*F)
X = np.dot(linalg.inv(K - w**2 * M), F)

print '\nIn this case, the magnitude of oscillation of the first mass is {}m.'.format(X[0])
print 'The magnitude of oscillation of the second mass is {}m.\n'.format(X[1])
```

```python
# Now, look at the "joined" case... the system reduces to a 1DOF mass-spring system
m = m1 + m2               # The new total mass (kg)

wn = np.sqrt(k1 / m)      # The natrual frquency of the 1DOF system

one_dof_amp = F1 / (m * (wn**2 - w**2))

print '\nThe amplitude of oscillation if the two masses are rigidly joined is {}m.\n'.format(one_dof_amp)
```

Looking at the two amplitudes, mass $m_2$ will have the same amplitude oscillation if used as a vibration absorber or rigidly attached to mass $m_1$.<br><br>

## Problem 4.68
In this problem, we can check that the only time that mass $m_3$ remains stationary, $m_2$ is also stationary.

```python
m1 = 3.0
m2 = 2.0
m3 = 1.0

k1 = 4.0
k2 = 4.0
k3 = 4.0

M = np.array([[m1, 0,  0],
              [0, m2,  0],
              [0,  0, m3]])

K = np.array([[ k1,      -k1,   0],
              [-k1, k1 + k2, -k3],
              [  0,     -k3,  k3]])

F2 = 1.0

F = np.array([0, F2, 0])

w = np.linspace(0.01,4,4000)
X = np.zeros((len(w),3))

# This is (K-w^2 M)^-1 * (w^2*F)
for ii, freq in enumerate(w):
    inverse = linalg.inv(K - freq**2 * M)
    X[ii,:] = np.dot(inverse, F)

# Let's mask the discontinuity, so it isn't plotted
pos = np.where(np.abs(X[:,0]) >= 10.0)
X[pos,:] = np.nan
w[pos] = np.nan
```

```python
# Set the plot size - 3x2 aspect ratio is best
fig = figure(figsize=(6,4))
ax = gca()
subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Change the axis units to serif
setp(ax.get_ymajorticklabels(),family='serif',fontsize=18)
setp(ax.get_xmajorticklabels(),family='serif',fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
xlabel('Input Frequency (rad/s)',family='serif',fontsize=22,weight='bold',labelpad=5)
ylabel('Vibration Amplitude',family='serif',fontsize=22,weight='bold',labelpad=10)

plot(w, np.abs(X[:,0]),linewidth=2,label=r'$x_1$')
plot(w, np.abs(X[:,1]),linewidth=2,linestyle="--",label=r'$x_2$')
plot(w, np.abs(X[:,2]),linewidth=2,linestyle="-.",label=r'$x_3$')

# uncomment below and set limits if needed
# xlim(0,2.5)
ylim(0,2.5)

# Create the legend, then fix the fontsize
leg = legend(loc='upper right', fancybox=True)
ltext  = leg.get_texts()
setp(ltext,family='serif',fontsize=18)

# Adjust the page layout filling the page using the new tight_layout command
tight_layout(pad=0.5)

# save the figure as a high-res pdf in the current folder
# It's saved at the original 6x4 size
# savefig('Problem4p68.pdf')

fig.set_size_inches(9,6) # Resize the figure for better display in the notebook
```

## Problem 4.75

```python
# Let's improve the printing of NumPy arrays for these eigenvalue/eigenvector problems
np.set_printoptions(precision=3, suppress=True)
```

```python
# Define the matrices
m1 = 1.0
m2 = 5.0

k1 = 100.0 
k2 = 200.0
k3 = 50.0

F = [1.8318, -1.3145]

M = np.asarray([[m1, 0],
                [0,  m2]])

K = np.asarray([[k1 + k2, -k2],
                [-k2,      k2 + k3]])
```

```python
# We'll use the scipy version of the linear algebra
from scipy import linalg

eigenvals, eigenvects = linalg.eigh(K,M)
```

```python
print '\n'
print 'The resulting eigenalues are {:.2f} and {:.2f}.'.format(eigenvals[0], eigenvals[1])
print '\n'
print 'So the two natrual frequencies are {:.2f}rad/s and {:.2f}rad/s.'.format(np.sqrt(eigenvals[0]), np.sqrt(eigenvals[1]))
print '\n'

print 'The first eigenvector is ' + str(eigenvects[:,0]) + '.'
print '\n'
print 'The second eigenvector is ' + str(eigenvects[:,1]) + '.'
print '\n'
```

Now, let's use these to form the normal form, and check if we get what we expect (decoupled equations).

```python
K_prime = np.dot(np.dot(eigenvects.transpose(),K), eigenvects)
M_prime = np.dot(np.dot(eigenvects.transpose(),M), eigenvects)
F_prime = np.dot(eigenvects.transpose(), F)

print 'The de-coupled M matrix is:'
print '{}.'.format(M_prime)

print '\nThe de-coupled K matrix is:'
print '{}.'.format(K_prime)

print '\nThe de-coupled F matrix is:'
print '{}.'.format(F_prime)
```

From this, we can see that the de-coupled equation of motion is:

$ \quad \ddot{H} + \begin{bmatrix}\omega_1^2 & 0 \\ 0 & \omega_2^2\end{bmatrix}H = \begin{bmatrix}0 \\ -1.924\end{bmatrix}\cos{10t}$

This shows that the forcing function exactly matches the second mode.

## Problem 4.96

```python
m1 = 1.0
m2 = 1.0
m3 = 1.0
m4 = 1.0

k1 = 4.0
k2 = 4.0
k3 = 4.0
k4 = 4.0
k5 = 4.0

M = np.array([[m1, 0,  0,  0],
              [0, m2,  0,  0],
              [0,  0, m3,  0],
              [0,  0,  0, m4]])

K = np.array([[k2 + k3,      -k2,      -k3,       0],
              [    -k2,  k2 + k4,        0,     -k4],
              [    -k3,        0,  k3 + k5,     -k5],
              [      0,      -k4,      -k5, k4 + k5]])

eigenvals, eigenvects = linalg.eigh(K,M)
```

```python
print '\n'
print 'The resulting eigenalues are {:.2f}, {:.2f}, {:.2f}, and {:.2f}.'.format(eigenvals[0], eigenvals[1], eigenvals[2], eigenvals[3])
print '\n'
print 'So the natrual frequencies are {:.2f}rad/s, {:.2f}rad/s, {:.2f}rad/s, and {:.2f}rad/s.'.format(np.sqrt(eigenvals[0]), np.sqrt(eigenvals[1]), np.sqrt(eigenvals[2]), np.sqrt(eigenvals[3]))
print '\n'

print 'The first eigenvector is ' + str(eigenvects[:,0]) + '.'
print '\n'
print 'The second eigenvector is ' + str(eigenvects[:,1]) + '.'
print '\n'
print 'The third eigenvector is ' + str(eigenvects[:,2]) + '.'
print '\n'
print 'The fourth eigenvector is ' + str(eigenvects[:,3]) + '.'
print '\n'
```
